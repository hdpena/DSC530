```python
def EvalCdf(sample, x):
    count = 0.0
    for value in sample:
        if value <= x:
            count += 1

    prob = count / len(sample)
    return prob
```

Digram 1.12: Barchart Showing Crime Codes Commited By Area Name

```python
crimeByArea = crime_data['Area Name'].value_counts().sort_index()
crimeCommonType = {}
for area in crimeByArea.keys():
    crimeArea = crime_data[crime_data['Area Name'] == area]['Crime Code Description'].value_counts()[:10]
    for crType in crimeArea.keys():
        if not crType in crimeCommonType:
            crimeCommonType[crType] = [area]
        else:
            crimeCommonType[crType].append(area)
    crimeArea = crimeArea.plot(kind = 'bar',title = "Crimes in " + area)
    plt.show()
```
